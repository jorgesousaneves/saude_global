```python
from pyspark.sql.functions import col
import requests
from io import StringIO
import pandas as pd # Usei panda apenas para ler a string e ser√° convertido para Spark imediatamente.

SOURCE_URL = "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv"
TABLE_BRONZE = "saude_global.bronze.owid_covid_data_raw"

response = requests.get(SOURCE_URL)
response.raise_for_status() 

csv_content = StringIO(response.text)
df_pandas = pd.read_csv(csv_content)

df_raw = spark.createDataFrame(df_pandas)

df_raw.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(TABLE_BRONZE) 
```
